### Analysis on MTA Turnstile Data

The goal of this project is find out the best times for Zog Sports to have team members in the field to advertise our upcoming summer softball league. The goal is to find the times with the most foot traffic, or in other words the times with the most entries and exists.

In order to answer this question I took a look at one of the stations we were interetsed in: Yankee Stadium. I grouped the data by date and time and was able to calculate the total number of people who entered the station each day in each 4 hour interval. I then pulled up the Yankees home schedule and found 47 gamedays. Those are the days I was interested in. The above bar graph shows the hours of the day that recorded the most entries for game days specifically. With the above information we can start to shape the conclusion that from around 12:00 (before day games) and 4:00 (after day games have ended) would likely be the best time to put our team members in the field. 

Next steps for this project is to continue to plot this data for more stops around Manhatten. This will include stops closer to Central Park.
